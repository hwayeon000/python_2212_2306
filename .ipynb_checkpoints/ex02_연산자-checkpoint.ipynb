# 3)연산자
- 산술연산자
    - / : 나누기를 의미 (소수점까지 표시)
    - // : 몫 (소수점 이하 버림)
    - % : 나머지 (나누고 남은 값)

```python
num1, num2 = 10, 7

print(num1/num2)
print(num1//num2)
print(num1%num2)
```

```python
# 문자열 더하기 -> 가능
s1 = "안녕!"
s2 = "❤❤"
print(s1 + s2)
```

```python
#  문자열과 숫자의 연산 --> 지원하지 않는다
num3, s3 = 10, '7'
print(type(num3))
print(type(s3))
# type : 변수에 담긴 데이터 자료형을 확인하는 함수

print(num3, s3)
```

- 형변환
    - 문자와 숫자 더할 수 없기 때문에 자료형을 통일해 줄 때 사용
    - 변환할 자료형(변수명)

```python
# 문자열을 숫자로
# 변경하고싶은 자료형(변수명)
int(s3)
```

```python
print(type(s3))
print(type(int(s3)))
```

```python
# 숫자를 문자로
print(type(str(num3)))
```

```python
# 더하기 진행
print(num3 + int(s3))
print(str(num3) + s3)
```

## input


##### python 에서 입력받는 방법
- input() 함수
- 주의점! 입력받은 데이터는 무조건 문자열로 입력된다!

```python
num = input()
print(type(num))
```

```python
u_num = input()
# 정수타입으로 형변환
int(u_num)
print(type(u_num))
```

```python
num1, num2 = map(int, input().split())
print(num1)
print(num2)
```

```python
num1, num2 = map(int, input("숫자 두개를 띄어서 입력해주세요>>").split())
```

```python
print(f"num1은 {num1}, num2는 {num2}")
print(f"더하기 결과 : {num1+num2}")
print(f"빼기 결과 : {num1-num2}")
print(f"곱하기 결과 : {num1*num2}")
print(f"나누기 결과 : {num1/num2}")
```

```python
time = int(input("시간 입력 >> "))
hour = time//3600
minute = (time%3600)//60
second = time%60
print(f"{hour}시간 {minute}분 {second}초")
```

- 지수연산자
    - 제곱을 계산해주는 연산자 (**)

```python
# 2의 4 제곱
print(2**4)
```

- 대입연산자
    - = (대입연산자)
    - +=, -=, *= ... (복합대입연산자)

```python
a = 10
b = 3
a *= b   # a = a * b
print(a)

# 파이썬에는 ++ 연산자 없음 주의!
```

- 비교연산자
    - 같다 : ==
    - 다르다 : !=
    - <, <=, >, >= : 꺽새 방향이 열린쪽이 더 크다
    - 결과값이 논리형 ( boolean: True, False ) 형태의 자료

```python
print(3 == 4)
```

```python
a = True
b = 0
print(not a, not b)
```

- 논리연산자
    - 논리형(True, False)을 연산해주는 연산자
    - not : 논리를 부정해주는 연산자 (논리 앞에 써주기) 
    - and : 둘다 True 일 때만 True라는 결과값 반환
    - or :  둘 중 하나라도 True가 있다면 True는 결과값 반환
    - 주의 :  논리형 데이터끼리 연산을 해 줘야 한다!

```python
a, b = 10, 7
print(a!=b and a==b)
print(a!=b or a==b)
print(not a!=b)
```

- 삼항연산자
- 조건식의 결과가 참이면 a를, 거짓이면 b를 반환
- 간단한 조건문을 쓸 때 사용

```python
"맞다" if 10>3 else "틀리다"
```

```python
score = 80
"합격" if score > 60 else "불합격"
```

```python
age = 19
"성인 입니다" if age > 19 else "미성년자 입니다"
```

```python
userNum = int(input("정수입력 >>"))
result = "짝수입니다." if userNum % 2 == 0 else "홀수입니다."
print(result)
```
