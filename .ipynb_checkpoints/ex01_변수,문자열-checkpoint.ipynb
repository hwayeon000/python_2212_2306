### 문자열 인덱싱 슬라이싱
- 인덱싱 : 문자열 내부에서 내가 원하는 데이터 하나를 뽑아 올 때 사용
- 슬라이싱 : 문자열 내부에서 범위를 주고 데이터들을 잘라올 때 사용

```python
s6 = "My name is YH"
s6
```

```python
# 인덱싱 방법 : 변수명[인덱스 번호]
# 'M' 인덱싱 하기
print(s6[0])
# 인덱스 0 부터 시작
```

```python
# i 가져오기
print(s6[8])
print(s6[-5])
# 띄어쓰기도 인덱스 번호를 부여 받는다~
```

```python
# H 가져오기
print(s6[-1])
print(s6[12])
# 음수를 활용해서 인덱싱 가능 (뒤에서부터 count 한다)
```

```python
# 슬라이싱 : 여러개의 데이터를 가져올 때 사용
# 슬라이싱 방법 : 변수명[시작인덱스:끝인덱스+1:스텝]
# My 가져오기
print(s6[0:2])
```

```python
# ;'is YH'가져오기
print(s6[8:13])
print(s6[-5:])
# 콜론 뒤 인덱스를 비워놓으면 끝까지 가져오겠다
```

```python
# 'My name'
print(s6[:7])
# 콜론 앞 인덱스를 비워놓으면 처음부터 가져오겠다
```

```python
# 모든 문자 가져오려면
# 앞 뒤 모두  비우고 콜론만 작성
print(s6[:])
```

- 포매팅이란 문자열 내에 특정 변수를 삽입하여 문자를 출력할 때 사용

### 출력포매팅
- 3가지 방법
    - % 기호 포매팅
    - format 함수포매팅
    - f문자열 포매팅

- %기호 포매팅
    - 주의1 % 기호 개수만큼 변수를 적어줘야 함!!
    - 주의2 문자열 내부 % 기호 순서에 맞게 변수 순서 배치
    
    - 단점 : 포맷코드를 외워야 함

```python
# %기호 포매팅 - % 사용하여 데이터 사용
month = 2
day = 9
str1 = "내일은"
print("%s %d월 %d일 입니다."%(str1, month, day))
```

```python
# format 함수를 사용한 포매팅
month = 2
day = "구"
s8 = "오늘은 {}월 {}일 입니다.".format(month, day)
print(s8)
```

```python
# f 문자열 포매팅

s9 = f"오늘은 {month}월 {day}일 입니다."
print(s9)
# 장점 : 직관적이다 -> 어떤 변수가 어디에 들어가는 지 직관적으로 확인 가능
```

```python
x, y = 100, 200
add = x + y

ans1 = "%d와 %d의 합은 %d입니다."%(x, y, add)
ans2 = "{}와 {}의 합은 {}입니다.".format(x, y, add)
ans3 = f"{x}와 {y}의 합은 {add}입니다."
print(ans1)
print(ans2)
print(ans3)
```

### 문자열 함수 사용해보기

```python
#  count 함수 -> 특정 문자 몇개 들어가는지 세어준다
# 변수명.count('문자')
name = "My name is hwayeon"
print(name.count('m')) # 대소문자 구별!
print(name.count('yeon')) # 단어도 가능!
```

```python
# find 함수 -> 문자 인덱스 반환
print(name.find('n'))
print(name.find('a')) # 왼쪽부터 검색하여 가장 먼저 만자는 앞 인덱스 반환
print(name.find('A')) # 문자열에 없는 문자는 '-1'을 반환
```

```python
# index함수 -> 문자 인덱스 반환
print(name.index('n'))
print(name.index('a'))
# print(name.index('A')) -> Error
# 인덱스 함수는 없는 문자가 있으면 오류!
```

```python
# replace 함수 -> 문자열 바꾸기
# 변수명.replace("변경전문자", "변경후문자")
print(name.replace("My", "💘"))
print(name)
# 이모지 : win + .
```
