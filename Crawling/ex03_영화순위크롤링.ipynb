```python
# 라이브러리 불러오기
# 페이지 정보를 요청, 응답받는 라이브러리
import requests as req
# 컴퓨터가 알 수 있는 html언어로 변경해주는 라이브러리
from bs4 import BeautifulSoup as bs
```

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"
res = req.get(url)
res
```

```python
# 객체를 문자열로 가져옴
res.text
# 가져온 문자열을 html 객체로 파싱?
# 뷰티풀숲 객체화
soup = bs(res.text,"lxml")
soup

data = soup.select("div.tit3>a")
# select로 가져온 데이터 결과는 리스트 타입으로 저장
# .text를 하게 되면 오류 발행! .text는 요소에만 해야 한다!!
# 오류가 나지 않게 하려면 인덱싱을 통해 요소 하나만을 뽑아 .text 진행~
data[:5]

# 컨텐츠 추출
data[0].text
```

```python
# 반복문 for문을 활용하여 모든 데이터의 컨텐츠만 추출
for i in data:
    print(i.text)
```

```python
# 개수가 같은 여러개의 데이터 하나의 for문을 사용하여 추출
title_list = []
for i in range(len(data)):
    title_list.append(data[i].text)

title_list
```

### 영화제목&평점데이터 수집하기

```python
# 페이지 정보 요청 후 응답받기
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20230301"
res = req.get(url)
res.text
# 문자열 -> html 변경
soup = bs(res.text,"lxml")

title = soup.select("div.tit5>a")
title[0]
```

```python
grade = soup.select("td.point")
grade[0]
```

```python
print(len(grade))
print(len(title))
```

```python
title_list = []
grade_list = []
index_list = []
for i in range(len(title)):
    title_list.append(title[i].text)
    grade_list.append(grade[i].text)
    index_list.append(i+1)
    print("제목 :" + title[i].text + ", 평점 :" + grade[i].text)
```

```python
import pandas as pd
# mData = {"영화제목":title_list,"평점":grade_list}
# title_list
# grade_list
# mDf = pd.DataFrame(mData, index=index_list)
# mDf.index.name = "순위"

# 쌤꺼
m1Data = {"영화제목":title_list,"평점":grade_list, "순위":index_list}
mDf = pd.DataFrame(m1Data, index="순위")
# mDf.set_index("순위", inplace=True)
```
