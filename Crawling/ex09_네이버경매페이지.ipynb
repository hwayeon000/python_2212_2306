```python
# 웹브라우저 "통제"
from selenium import webdriver as wb
# 컴퓨터용 키보드
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup as bs
# 시간 지연
import time
# 로딩바
from tqdm import tqdm_notebook as tq
# 이미지 주소 파일로 변환해주는 라이브러리
from urllib.request import urlretrieve
# 폴더 생성/삭제해주는 라이브러리(파일 시스템 라이브러리)
import os
```

```python
url = "https://auction1.land.naver.com/auction/ca_list.php"
driver = wb.Chrome()
driver.get(url)
time.sleep(1)

num_list = []
type_list = []
add_list = []
price1_list = []
price2_list = []
status_list = []
date_list = []

cnt=0
while cnt<3:
    try:
        cnt += 1
        soup = bs(driver.page_source,"lxml")

        nums = soup.select("td.num")
        types = soup.select("td.num+td")
        adds = soup.select("td.area>a")
        price1 = soup.select("p.num_type1>strong")
        price2 = soup.select("p.num_type2>strong")
        status = soup.select("td.price+td")
        dates = soup.select("td.date")

        for i in range(len(nums)):
            num_list.append(nums[i].text)
            type_list.append(types[i].text)
            add_list.append(adds[i].text)
            price1_list.append(price1[i].text)
            price2_list.append(price2[i].text)
            status_list.append(status[i].text)
            date_list.append(dates[i].text)

        next_btn = driver.find_element(By.CSS_SELECTOR,"div.pagnavi>strong+a")
        next_btn.click()
        time.sleep(2)
        
    except:
        print("모든페이지를 클릭하셨습니다.")
        break

driver.quit()
```

```python
dic = {"사건번호":num_list,
       "물건종류":type_list,
       "소재지":add_list,
       "감정가":price1_list,
       "최저입찰가":price2_list,
       "물건상태":status_list,
       "매각기일":date_list
       }
dic

import pandas as pd
nData = pd.DataFrame(dic)
```
