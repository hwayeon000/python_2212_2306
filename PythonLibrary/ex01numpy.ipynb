### Numpy
- numerical python의 약자
- 파이썬 내부에서 데이터의 숫자 계산을 간편하게 해주는 라이브러리
- 파이썬 자료형인 list와 비슷하지만, "동일한 자료형"을 담는다는 차이점이 있다!
- 반복문 없이 각 데이터의 연산이 가능하다
- 각 값들은 index 번호를 부여받는다
- ndarray 라고 불리는 넘파이 배열 생성가능

```python
# numpy 라이브러리 불러오기
import numpy as np
```

```python
# 1차원 배열 생성
# np가 가지고 있는 기능중 배열(array)을 생성할 수 있는 기능이 있다

list1 = [1,2,3,4,5]
print(list1, type(list1))
print(np.array(list1), type(np.array(list1)))
arr1 = np.array(list1)
# 리스트형대의 데이터 앞에 array라고 소괄호가 붙은 상태가 되면 "넘파이 배열"이라고 한다
```

### 이차원 넘파이배열 생성하기!

```python
# 가로축 : 행 데이터
# 세로축 : 열 데이터
# 2행 3열의 데이터

arr2 = np.array([[1,2,3], [4,5,6]])
print(arr2)
```

### ndarray(넘파이 배열)의 속성을 활용하여 자료 구조 확인하기

- 배열의 크기 확인 : shape

```python
# 배열명.shape

# 1차원 : (데이터의 개수,)
print(arr1.shape)
# 2차원 : (행, 열)
print(arr2.shape)
```

- 배열의 요소 개수 확인하기 : size


```python
# 배열명.size ==> 데이터의 총 개수 확인
print(arr1.size)
print(arr2.size)
```

- 배열의 차원 확인하기 : ndim

```python
# 데이터 차원을 알려준다
print(arr1.ndim)
print(arr2.ndim)
```

- 배열의 타입 확인 : dtype

```python
# 배열명.dtype
print(arr1.dtype)
print(arr2.dtype)
```

```python
arr3 = np.array([[[1,2],[1,2]],[[4,5],[4,5]]])
print("배열의 차원 :", arr3.ndim)
print("배열의 타입 :",arr3.dtype)
print("배열의 크기 :", arr3.shape)
print("배열의 개수 :", arr3.size)
```

### 특정한 방식으로 넘파이 배열 생성하기!

```python
# 3가지
# 1. 배열안에 모든 데이터 초기화 후 생성하는 방법
# np.zeros(배열의 크기 --> 튜플 형태로 작성)
arr_zero = np.zeros((2,2))   #튜플 형대로 행과 열 개수 지정 (2차원 배열)
print(arr_zero)
```

```python
# 2. 배열 안의 모든 데이터 1로 초기화하여 생성
# np.ones(배열의 크기)
arr_one1 = np.ones((5))
print(arr_one1)   # 데이터를 하나만 쓰면 1차원 배열 생성 (숫자 == 데이터의 개수)
arr_one = np.ones((3,3))
print(arr_one)
```

```python
# 내가 원하는 특정값으로 초가화 해서 배열 생성
# np.full((배열의 크기), 특정 값)  //2개의 매개변수
arr_full = np.full((2,3), 7)
print(arr_full)
```

### array 연산
- numpy는 기본적으로 array간의 사직연산 지원!!
- 같은 위치에 있는 요소들끼리 연산

```python
arr_a = np.array([1,2,3,])
print(arr_a)
arr_b = np.array([4,5,6,])
print(arr_b)
```

```python
print(arr_a + arr_b)
print(arr_a * arr_b)
```

- 이차원배열의 연산

```python
arr2_a = np.array([[1,2,3], [4,5,6]])
print(arr2_a)
arr2_b = np.array([[4,5,6],[7,8,9]])
print(arr2_b)
```

```python
print(arr2_a + arr2_b)
print(arr2_a * arr2_b)

# numpy 배열에 정수를 연산하면 각 인덱스에 모두 연산 진행
print(arr2_a * 3)
```

## Boolean 색인/인덱싱
- boolean : 논리 자료형
- 배열 안에서 조건을 충족하는 True인 값들만 추출해주는 인덱싱 방법

```python
score = np.array([80,75,55,96,30])
name = np.array(["민지","유빈","진영","고은"])

print(score, name)

# 현재는 논리연산만 진행한 결과
print(score>=80)
```

```python
# 배열[조건]
print(score[score>=80])
# 이름이 '유빈'인 사람만 추출
print(name[name=="유빈"])
```

```python
score2 = np.array([[60,60],[70,80],[90,20],[50,50]])
print(score2)

# "유빈"의 성적 가져오기
print(score2[name=="유빈"])
```

### numpy 범용함수
- 범용함수란? 널리 퍼져있는 수학 함수
- sum(), mean(), max(), min()

```python
# arr이란 변수에 래덤수 (1~10) 들어있는 2행 5열의 배열
arr = np.random.randint(1, 11, (2,5))
print(arr)

# np.sum() : 합계
print(sum(arr))
print(np.sum(arr))
# np.mean() : 전체 요소값의 평균
print(np.mean(arr))

# 제곱근 구하는 함수
# np.sqrt(변수명)
print(np.sqrt(arr))


# max()
print(np.max(arr))
# # min()
print(np.min(arr))
```

```python
arr_abs = np.array([-1, 2, -3, 4, -5])

# 절대값 구하는 함수
# abs()
print(abs(arr_abs))
```
