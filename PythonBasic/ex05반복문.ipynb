### 반복문




### 1) while
- 반복의 횟수가 명확하지 않을 때 사용하는 반복문
- 조건식이 True일 경우 실행문장을 실행
- while 조건식 :
 (들여쓰기)실행문장

```python
num = 13
while num :
    print("파이썬 최고")
    num -= 1
```

```python
# 두 개의 정수를 입력받아 더하는 코드
# 단, 두개의 정수가 모두 0이 입력되면 종료

while True:
    num1, num2 = map(int, input("두 정수 입력>>").split())
    print(f"첫번째 정수 : {num1}, 두번째 정수 {num2}")
    if num1 == 0 and num2 == 0:
        print("프로그램 종료")
        break
    print(f"두 정수의 합 : {num1+num2}")
    
```

```python
num = 13
while num :
    num -= 1
    if num%2==0:
        continue
    print(num)
```

```python
# 현재 몸무게와 목표 몸무게
nowWt = int(input("현재 몸무게 >>"))
goalWt = int(input("목표 몸무게 >>"))
print(f"현재 몸무게 : {nowWt}")
print(f"목표 몸무게 : {goalWt}")
i = 0
while True:
    i+=1
    minWt = int(input("감량 몸무게 : "))
    print(f"{i}주차 감량 몸무게 :", minWt)
    nowWt -= minWt
    if nowWt <= goalWt:
        print(f"{nowWt}kg 달성!! 축하합니다!")
        break
    
```

##### 랜덤수 추출하는 방법
- random 라이브러리 활용

```python
# 라이브러리 불러오기
import random as rd

# 랜덤 추 추출(1~10 범위의 랜덤 정수 추출)
num = rd.randint(1, 10)
print(num)
```

```python
num = rd.randint(1,50)

while True:
    myNum = int(input("숫자를 입력 >>"))
    res = "큰 수" if num>myNum else "작은 수"
    print(f"{myNum}보다 {res}입니다.")
    if myNum == num:
        print("정답!!")
        break
```

#### for문
- 명확하게 반복의 횟수가 정해져 있을 때 사용

- for 변수 in 문자열(or 리스트 or 튜플 or range()):
- (들여쓰기)실행문장

```python
s1 = "헤이 모두들 안녕"
for i in s1:
    print(i)
```

```python
# 개행 없이 출력해보기 => end 속성 활용
s2 = "오늘은 ✨영민씌✨ 생일!✨"
for i in s2:
    print(i, end="")
# 속성 보기 : shift + tab
```

```python
score_list = [90,45,70,60,55,70]

cnt = 1
for i in score_list:
    res = "합격" if i >= 60 else "불합격"
    print(f"{cnt}번 학생은 {res}입니다")
    cnt += 1

# 인덱스 번호로 가져오면 중복되는 점수가 있을 시 앞에 있는 인덱스를 가져온다!! 
# 참고만 하쟈..
# for i in score_list:
#     if i >= 60:
#         print(f"{score_list.index(i)+1}번 학생은 합격입니다")
#     else:
#         print(f"{score_list.index(i)+1}번 학생은 불합격입니다")
#     cnt += 1
```

#### for문과 힘께 사용되는 range() 함수
- range() 함수는 필요한 숫자를 만들어 내는 기능
- for문과 함께 사용 - 반복 횟수를 정할 때
- for i in range(시작, 마지막숫자+1, 스텝)



```python
# 1 ~ 10까지 출력
for i in range(11):
    print(i, end=" ")
```

```python
# 97 ~ 77까지 출력
for i in range(97, 76, -1):
    print(i, end=" ")
# 줄바꿈
print()
# 23 ~ 40까지 출력
for i in range(23, 41):
    print(i, end=" ")
```

```python
# 두 수를 입력받아 반복문 출력
start, end = map(int, input("정수 입력").split())
print(f"내가 입력한 수 {start},{end}")
for i in range(start, end+1):
    print(i, end=" ")
```

```python

num = 0
for i in range(1, 101):
    if i % 3 == 0:
        num += i
print(num)


# 🔴🟠🟡 중요 🟡🟠🔴
for i in range(1, 10):
    print(f"2 * {i} = {2 * i}")

```

```python
for j in range(2, 4):
    print(f"{j}단")
    for i in range(1, 10):
        print(f"{j} * {i} = {j * i}")
```

```python
num = int(input("정수 입력>>"))
print("입력값 :", num)
print(f"{num}의 약수 : ", end="")
for i in range(1, num+1):
    if num % i == 0:
        print(i, end=" ")
```

##### 2. 리뷰 길이의 평균값

```python
num = 0
for i in data:
    num += len(i[0])
print(num)
print(num/len(data))
print(f"리뷰의 평균 길이는 '{num/len(data)}'입니다.")
```

##### 3. 긍정리뷰/부정리뷰 개수 구하기
- 긍정 1, 부정 0 : count()
- 긍정 2499 / 부정 2501

```python
a = []
for i in data:
    a.append(i[1])

p = a.count(1)
n = a.count(0)
print()
print(f"긍정 : '{p}'개\n부정 : '{n}'개")
```

```python
po = 0
na = 0
for i in data:
    if i[1] == 1:
        po += 1
    else :
        na += 1

print()
print(f"긍정 : '{po}'개\n부정 : '{na}'개")
```

```python
po1 = 0
na1 = 0
for i in data:
    if i[1]:
        po1+=1
na1 = len(data)-po1
print()
print(f"긍정 : '{po1}'개\n부정 : '{na1}'개")
```
