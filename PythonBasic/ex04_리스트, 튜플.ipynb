### in, not in
- in :   (x)   True
- not in :   (x)    True


```python
myList = ["", "", "", ""]

print("" in myList)
print("" in myList)
print("" not in myList)
```

```python
lst = ["", "", "", "", ""]
```

```python
print(""*6)
myFt = input("   >>")
print(f"   >> {myFt}")

if myFt in lst :
    myFtIndex = lst.index(myFt)
    print(f"{myFt}()  {myFtIndex}  .")
else:
    print(f"{myFt}()   .")
print(""*6)
```

```python
#       
#    

# =================================================================

#      : count()
#  : .count()
```

```python
name = "Hi, My name is Younghwa"
print(""*6)
search = input("   >>")
print(f"   >> {search}")

if search in name :
    myFtIndex = name.index(search)
    countL = name.count(search)
    print(f"{search}()  {countL} .")
else:
    print(f"{search}()   .")
print(""*6)
```
