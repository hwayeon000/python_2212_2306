# 조건문
- 조건에 따라 실행 흐름을 다르게 하는 문법 

### if문
- if 조건식(논리형데이터/논리형결과):
- (들여쓰기) 실행문장
- 
- 조건식이 True일 경우 실행문장 실행


```python
if True:
    print("실행")
```

```python
if False:
    print("실행")
    # if문은 조건식이 False 였을 때 아무것도 실행하지 않고 끝남
```

```python
# 들여쓰기중요성!!!!
if False:
    print("들여쓰기 한 출력문") # 1번
print("들여쓰기 하지 않은 출력문") # 2번
# 들여쓰기를 하지 않은 2번 문장만 출력
# if문의 영향을 벗어나고 싶다면 들여쓰기 없애주기
```

```python
# 조건문 if 활용해서 money가 10000원 이상이면 "택시를 탄다" 반환
money = 11000
# money = 9000
if money >= 10000:
    print("택시를 탄타")
```

### if-else문
    - else는 참이 아닌모든 경우를 포함하기 떄문에 조건식을 적어 줄 필요 없음


```python
if False:
    print("실행문장 1")
else:
    print("실행문장 2")
# else 는 if와 같은 선상에 적어주기 (자동 들여쓰기 되기 때문에 유의!)
```

```python
money = 9000
print("현재 내 자산 : ", money)
if money >= 10000:
    print("택시를 탄다")
else:
    print("버스를 탄다")
```

```python
num = int(input("정수 입력>>"))

print("입력값", num)
if num % 3 == 0 and num % 5 == 0:
    print("3과 5의 배수입니다.")
else:
    print("3과 5의 배수가 아닙니다.")
```

### if-elif-else 문
    - 조건이 여러개일 때


```python
# 돈이 만원 이상이면 택시, 5000이상이면 버스 5000미만이면 걷기;

money = 9000
print("현재 내 자산 : ", money)
if money >= 10000:
    print("택시를 탄다")
elif money >=5000:
    print("버스를 탄다")
else:
    print("걸어간다")
```

```python
num1, num2 = map(int, input("정수 입력 >>").split())

# 3개의 조건을 판별하여 실행
if num1>num2:
    print("첫 번째 정수가 더 큽니다.")
elif num1<num2:
    print("두 번째 정수가 더 큽니다.")
else:
    print("두 수가 똑같습니다..")

```

```python
score = int(input("정수 입력 >>"))
print("내 점수",  score)

if 100 >= score >= 90:
    res = "A"
elif score >= 80:
    res = "B"
elif score >= 70:
    res = "C"
elif score >= 40:
    res = "D"
else:
    res = "F"
print(f"{score}점은 {res}학점입니다.")
```
