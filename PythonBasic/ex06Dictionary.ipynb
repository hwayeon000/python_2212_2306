## Dictionary 딕셔너리
- key와 value를 한 쌍으로 같는 자료형
- immutable(변할 수 없는) key와 mutable(변할 수 있는)한 value
- 순서가 없는 집합




- key : value 한 쌍으로 이루어진 구조
- 순서가 없는 형태
    - key : 변할 수 없는 데이터
    - value : 변할 수 있는 데이터 모두 가능
- {} 중괄호를 이용하여 생성

```python
myDic = {
    "name" : "ㅂㅎㅇ",
    "age" : 30,
    "height" : 160
}

print(type(myDic))
print(myDic)

# value 값 확인
# 딕셔너리명["key"]
print(myDic["name"])
```

```python
# value 값 변경
# 딕셔너리명[key값] = 변경 데이터
myDic["age"] = 25
print(myDic["age"])

# 새로운 데이터 추가 방법
# 추가할 값에 대한 key는 대괄호에 열어 적고, 대입연산자를 활용하여 값(value)을 추가
# 딕셔너리명["추가할 key값"] = 추가 할 value값
myDic["birth"] = 324
print(myDic)
```

```python
# dic 선언 및 초기화
dic_test = {
    "노래제목" : "파이팅 해야지"
}
# dic에 값 추가
dic_test["가수"] = "부석순"
dic_test["날짜"] = "2023.02.06"
print(dic_test)
```

```python
# 데이터 삭제하기
del dic_test["날짜"]

print(dic_test)
```

- 값(value) 가져오는 여러가지 방법

```python
# key값 대괄호 안에 적어 가져오기
print(dic_test["노래제목"])

# get 함수 사용해서 가져오기
# 딕셔너리명.get(key)
print(dic_test.get("노래제목"))
```

[key] vs get(key)
- 대괄호를 통해 가져오면 없는 key값을 대입시 오류가 나지만
- get 함수를 사용하여 가져오면 "None" 반환

```python
# [key] vs get(key)
# print(dic_test["나이"])  # 오류
print(dic_test.get("나이"))  # 오류가 나지 않고 None

# 대괄호를 통해 가져오면 없는 key값을 대입시 오류가 나지만
# get 함수를 사용하여 가져오면 "None" 반환
```

```python
# 딕셔너리 안에 있는 key 값만 가져오는 방법
print(dic_test.keys())
```

```python
# 딕셔너리 안에 있는 values 값만 가져오는 방법
print(dic_test.values())
```

```python
# .itmes() : key값과 values값을 튜플로 반환해주는 방법
print(dic_test.items())
```

```python
#딕셔너리와 반복문을 활용하는 방법
for key in dic_test.keys():
    print("key:", key)
for i in dic_test.keys():
    print("values:", dic_test[i])

# key 값과 values 값 둘 다 가져오는 방법
for key, values in dic_test.items():
    print("key:", key, ", values:" ,values)
```

```python
# vlaues 값에 또 다른 dic이 들어갈 수 있다
student = { "최영화" : { "age" : "20", "gender" : "여자" },
           "최유솔" : { "age" : "25", "gnender" : "남자" }}
```

```python
print(student["최유솔"]["age"])
print(student["최영화"]["gender"])
```

```python
# dic 데이터 안에 해당하는 데이터 존재하는지 여부 
# in, not in 연산자 : dic에 key값 유무 확인하는 연산자

print("최영화" in student)
print("age" in student["최영화"])
print("여자" in student["최영화"])
```

### 딕셔너리 최종 실습
- 각 과목별 점수 합계 구하기

```python
score_dic = { "조승재" : { "Python" : 100 , "Java" : 20 , "DataBase" : 50 },
             "주부은" : { "Java" : 100 , "Python" : 80 , "DataBase" : 89},
             "장영주" : { "DataBase" : 40 , "Java" : 100 , "Python" : 60 }
              }

print(score_dic.keys())
print(score_dic.values())
print(score_dic["장영주"])
```

```python
# 각 과목별 합계 
py = 0
ja = 0
db = 0
for key in score_dic.keys():
    for valKey in score_dic[key]:
        if valKey == "Python":
            py += score_dic[key]["Python"]
        elif valKey == "Java":
            ja += score_dic[key]["Java"]
        elif valKey == "DataBase":
            db += score_dic[key]["DataBase"]

print(f"Python 합계: {py}, Java 합계: {ja}, DataBase 합계 {db}")
```

```python
sb = score_dic['조승재'].keys()
for i in sb:
    temp = 0
    for j, k in score_dic.items():
        temp += score_dic[j][i]
    print(f"{i}합계: {temp} ", end="/ ")
```

```python
sb = score_dic['조승재'].keys()
p = 0
j = 0
d = 0
print(sb)
for key, val in score_dic.items():
    for k, v in val.items():
        if k == "Python":
            p += v
        elif k == "Java":
            j += v
        elif k == "DataBase":
            d += v
print(p, j, d)
```
