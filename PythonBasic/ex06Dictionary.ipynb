```python
score_dic = { "" : { "Python" : 100 , "Java" : 20 , "DataBase" : 50 },
             "" : { "Java" : 100 , "Python" : 80 , "DataBase" : 89},
             "" : { "DataBase" : 40 , "Java" : 100 , "Python" : 60 }
              }

print(score_dic.keys())
print(score_dic.values())
print(score_dic[""])
```

```python
#    
py = 0
ja = 0
db = 0
for key in score_dic.keys():
    for valKey in score_dic[key]:
        if valKey == "Python":
            py += score_dic[key]["Python"]
        elif valKey == "Java":
            ja += score_dic[key]["Java"]
        elif valKey == "DataBase":
            db += score_dic[key]["DataBase"]

print(f"Python : {py}, Java : {ja}, DataBase  {db}")
```

```python
sb = score_dic[''].keys()
for i in sb:
    temp = 0
    for j, k in score_dic.items():
        temp += score_dic[j][i]
    print(f"{i}: {temp} ", end="/ ")
```

```python
sb = score_dic[''].keys()
p = 0
j = 0
d = 0
print(sb)
for key, val in score_dic.items():
    for k, v in val.items():
        if k == "Python":
            p += v
        elif k == "Java":
            j += v
        elif k == "DataBase":
            d += v
print(p, j, d)
```
